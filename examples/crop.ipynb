```python
crinfo_auto = imtools.qmisc.crinfo_from_specific_data(segmentation, [5])
crinfo_auto
```

```python
data3d_cropped_auto = imtools.qmisc.crop(data3d, crinfo_auto)
plt.imshow(data3d_cropped_auto[30,:,:])
```

# Uncrop data

```python
data3d_uncropped = imtools.qmisc.uncrop(
    data3d_cropped_auto, 
    crinfo_auto, 
    [100, 100, 100])
plt.imshow(data3d_uncropped[30,:,:])
```
